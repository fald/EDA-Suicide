```python
desired_cols = ['suicides_per_100k', 'HDI', 'gdp_for_year ($)']
desired_info = ['mean', 'std', 'min', '25%', '50%', '75%', 'max', 'range']
# So we can add our range
description = data.describe()
description.loc['range'] = description.loc['max'] - description.loc['min']
description.loc[desired_info].iloc[:, 1:] # Bleh, needs int indices not just col names
#description.loc['max']
```

```python
# Various ways you may want to group data
# By age range (or generation), by country, by sex, by year
data_by_age = data.groupby('age')
data_by_country = data.groupby('country')
data_by_sex = data.groupby('sex')
data_by_year = data.groupby('year')
groups = [data_by_age, data_by_country, data_by_sex, data_by_year]
for group in groups:
    print(group.mean())
    print(group.median())
```

```python
# Not really correlated with time, but:
# ~5x male:female ratio, ouch.
for group in groups:
    print(group.agg(['mean', 'median']))
```

```python
# Scatter plots:
# Let's go with a simple case of suicides versus GDP - "Does money buy happiness"
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
year_start = 2016
year_end = 2020
subset = data[(data.year >= year_start) & (data.year <= year_end)]
pt1 = subset['gdp_per_capita ($)']
pt2 = subset['suicides/100k pop']
ax = plt.axes()
ax.scatter(pt1, pt2)
# I guess I should also label and stuff
ax.set(xlabel='GDP per capita ($)',
      ylabel="Suicides per 100k population",
      title="Suicide rates vs GDP")
```
